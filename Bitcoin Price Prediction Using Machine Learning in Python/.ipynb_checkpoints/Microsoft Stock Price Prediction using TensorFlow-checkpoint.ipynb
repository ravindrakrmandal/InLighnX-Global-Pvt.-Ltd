### To build a time-series forecasting model that predicts Microsoftâ€™s future stock prices using historical data and deep learning (LSTM).

### STEP 1: Importing Necessary Libraries and Dataset

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
df=pd.read_csv(r"C:\Users\RAVINDRA MANDAL\Desktop\Web Programing\Project\Inlight global pvt ltd\Microsoft Stock Price Prediction with Machine Learning\MicrosoftStock.csv")
```

### STEP 2: Data Preprocessing

```python
#Convert Date to DateTime & Set Index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.sort_index(inplace=True)
```

```python
#handle missing value
df.interpolate(method='linear', inplace=True)
```

```python
print(df.columns)
```

```python
#Simple Moving Average (SMA)
df['SMA_20'] = df['close'].rolling(window=20).mean()
```

```python
#Exponential Moving Average (EMA)
df['EMA_20'] = df['close'].ewm(span=20, adjust=False).mean()
```

```python
#Bollinger Bands
df['BB_upper'] = df['SMA_20'] + 2 * df['close'].rolling(20).std()
df['BB_lower'] = df['SMA_20'] - 2 * df['close'].rolling(20).std()
```

```python
#Relative Strength Index (RSI)
delta = df['close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(14).mean()

rs = gain / loss
df['RSI'] = 100 - (100 / (1 + rs))
```

```python
#drop none value
df.dropna(inplace=True)
```

```python
#feature selection and normilations
features = ['close', 'volume', 'SMA_20', 'EMA_20', 'BB_upper', 'BB_lower', 'RSI']
data = df[features]

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
```

```python
#train test split
train_size = int(len(scaled_data) * 0.8)

train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]
```

```python
#Create Time-Series Sequences
def create_sequences(data, seq_length=60):
    X, y = [], []
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i])
        y.append(data[i, 0])  # Close price
    return np.array(X), np.array(y)

SEQ_LEN = 60

X_train, y_train = create_sequences(train_data, SEQ_LEN)
X_test, y_test = create_sequences(test_data, SEQ_LEN)
```

### STEP 3: Exploratory Data Analysis (EDA)

```python
#stock price trend
plt.figure(figsize=(12,5))
plt.plot(df['close'], label='Close Price')
plt.title('Microsoft Stock Price Trend')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[features].corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

### STEP 4: Model Training and Selection

```python
#Build LSTM Model (TensorFlow)
model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(100),
    Dropout(0.2),
    Dense(1)
])
```

```python
#complie model
model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)

model.summary()
```

```python
#train model
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.1
)
```

### STEP 5: Model Evaluation and Prediction

```python
#predic
predictions = model.predict(X_test)
```

```python
# Create dummy array for inverse scaling
dummy = np.zeros((predictions.shape[0], scaled_data.shape[1]))
dummy[:, 0] = predictions[:, 0]

predicted_prices = scaler.inverse_transform(dummy)[:, 0]

actual_prices = scaler.inverse_transform(
    np.concatenate((y_test.reshape(-1,1), np.zeros((len(y_test), scaled_data.shape[1]-1))), axis=1)
)[:, 0]
```

```python
#Model Evaluation Metrics
mae = mean_absolute_error(actual_prices, predicted_prices)
rmse = np.sqrt(mean_squared_error(actual_prices, predicted_prices))
r2 = r2_score(actual_prices, predicted_prices)

print("MAE :", mae)
print("RMSE:", rmse)
print("RÂ²  :", r2)
```

```python
#actul vs predicted plot
plt.figure(figsize=(12,5))
plt.plot(actual_prices, label='Actual Price')
plt.plot(predicted_prices, label='Predicted Price')
plt.title('Actual vs Predicted Microsoft Stock Price')
plt.legend()
plt.show()
```

### next 30 day predicted

```python
future_steps = 30
last_sequence = scaled_data[-SEQ_LEN:].copy()

future_predictions = []

for _ in range(future_steps):
    # Predict next close price
    pred = model.predict(
        last_sequence.reshape(1, SEQ_LEN, scaled_data.shape[1]),
        verbose=0
    )

    # Store predicted close price
    future_predictions.append(pred[0, 0])

    # Create next input row
    next_row = np.zeros((scaled_data.shape[1],))
    next_row[0] = pred[0, 0]              # predicted Close
    next_row[1:] = last_sequence[-1, 1:]  # keep other features same

    # Slide window
    last_sequence = np.vstack((last_sequence[1:], next_row))
```

```python
dummy_future = np.zeros((future_steps, scaled_data.shape[1]))
dummy_future[:, 0] = future_predictions

future_prices = scaler.inverse_transform(dummy_future)[:, 0]

print("Next 30 Days Predicted Microsoft Stock Prices:")
print(future_prices)
```

```python
#Plot Future Forecast
plt.figure(figsize=(10,4))
plt.plot(future_prices, marker='o')
plt.title("Microsoft Stock Price Forecast (Next 30 Days)")
plt.xlabel("Days Ahead")
plt.ylabel("Price")
plt.show()
```
